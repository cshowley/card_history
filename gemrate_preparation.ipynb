```python
df = pd.read_csv('gemrate_pokemon_catalog_20260108.csv')
df.loc[df['UNIVERSAL_GEMRATE_ID'].notna(), "GEMRATE_ID"] = df.loc[df['UNIVERSAL_GEMRATE_ID'].notna(), "UNIVERSAL_GEMRATE_ID"]
df = df.drop_duplicates(subset=['GEMRATE_ID'])
df['YEAR'] = df['YEAR'].astype(str).str.split('-').str[0].str.strip()
df = df[["GEMRATE_ID", "YEAR", "SET_NAME", "NAME", "CARD_NUMBER", "PARALLEL"]]
df = df.fillna("")
df = df.astype(str)
df = df.reset_index(drop=True)
df.to_csv("gemrate_pokemon_db.csv", index=False)
```
